```python
import pandas as pd

# =====================================================
# LOAD INPUT CSV
# =====================================================

df_scores_sorted = pd.read_csv("Intensity/processing_intensity_scores.csv")

# =====================================================
# DEFINE CHEF CLASSIFICATION FUNCTION
# =====================================================

def assign_chef_class(pis):
    if pis >= 0.70:
        return "CHEF 5 – Highly Technical"
    elif pis >= 0.55:
        return "CHEF 4 – Moderately Technical"
    elif pis >= 0.40:
        return "CHEF 3 – Standard Processing"
    elif pis >= 0.25:
        return "CHEF 2 – Light Processing"
    elif pis >= 0.10:
        return "CHEF 1 – Very Basic"
    else:
        return "CHEF 0 – Minimal"

# =====================================================
# APPLY CLASSIFICATION
# =====================================================

df_scores_sorted["CHEF_Class"] = df_scores_sorted["Processing_Intensity"].apply(assign_chef_class)

# =====================================================
# SAVE OUTPUT CSV
# =====================================================

df_scores_sorted.to_csv("Intensity/processing_intensity_with_classes.csv", index=False)

# =====================================================
# SHOW FIRST 20 ROWS
# =====================================================

print(df_scores_sorted.head(20))
```
